```typescript
import * as d3 from "npm:d3";
import { createCanvas } from "https://deno.land/x/canvas/mod.ts"
import { display } from "https://deno.land/x/display@v0.1.1/mod.ts";

const canvas = createCanvas(200, 200);
const ctx = canvas.getContext("2d");

const data = [
    {element: "Logomark", semantic: 40, affective: 20, aesthetic: 40},
    {element: "Wordmark", semantic: 50, affective: 10, aesthetic: 40},
    {element: "Typography", semantic: 30, affective: 20, aesthetic: 50},
    {element: "Palette", semantic: 10, affective: 50, aesthetic: 40},
    {element: "Visual Motifs", semantic: 20, affective: 40, aesthetic: 40},
    {element: "Explanatory Illustrations", semantic: 50, affective: 10, aesthetic: 40},
    {element: "Narrative", semantic: 60, affective: 20, aesthetic: 20},
    {element: "Voice", semantic: 30, affective: 50, aesthetic: 20},
    {element: "Lexicon", semantic: 60, affective: 20, aesthetic: 20},
    {element: "Interactions", semantic: 30, affective: 40, aesthetic: 30}
  ];

const width = 200;
const height = 200;

const x = d3.scaleLinear().range([0, width]);
const y = d3.scaleLinear().range([height, 0]);

const color = d3.scaleOrdinal(d3.schemeCategory10);
ctx.translate(width / 2, height / 2);

// const svg = d3.create("canvas")
//   .attr("width", width)
//   .attr("height", height)
//   .append("g")
//   .attr("transform", `translate(${width / 2}, ${height / 2})`);

const points = ctx.selectAll("circle")
  .data(data)
  .enter()
  .append("circle")
  .attr("cx", d => x(d.semantic))
  .attr("cy", d => y(d.affective))
  .attr("r", 5)
  .style("fill", (d, i) => color(i));


// const xAxis = d3.axisBottom(x);
// const yAxis = d3.axisLeft(y);

// svg.append("g")
//   .attr("transform", `translate(0, ${height / 2})`)
//   .call(xAxis);

// svg.append("g")
//   .attr("transform", `translate(${width / 2}, 0)`)
//   .call(yAxis);

// ctx.drawImage(svg.node(), 0, 0);
display(canvas);




```
